# DEFINE *soup_a* as the a-tag with the req'd class label

```python
soup_a = soup.find_all('a',class_='v-align-middle')
```

```python
soup_a[0]
```

```python
soup_a[0].text
```

# for loop to generate LIST of repo_names as 'user/repo'

# R1 lang: R page:1

```python
repo_names = []
for i in range(0,len(soup_a)):
    a = soup_a[i].text
    repo_names.append(a)
```
